### Extracting Cluster Data

```python
import glob
import os
import re
import pandas as pd

# Specify the path to the folder containing the data
base_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/TumorInvasionParameterScan"
file_pattern = "**/cclc_math_path/data_*.txt"  # Use glob pattern for all txt files in nested subfolders

# Gather all file paths matching the pattern
file_paths = glob.glob(os.path.join(base_path, file_pattern), recursive=True)

# Lists to store extracted values
records = []

# Loop through each file, load data, and collect the required values
for file_path in file_paths:
    # Extract parameters from the filename using regex
    match = re.search(r'data_(-?[\d.]+)_([\d.]+)_([\d.]+)', os.path.basename(file_path))
    if match:
        # Extract parameter values
        contact_energy = float(match.group(1).rstrip('.'))
        chemotaxis_lambda = float(match.group(2).rstrip('.'))
        proliferative_probability = float(match.group(3).rstrip('.'))

        # Read the file content
        with open(file_path, 'r') as file:
            content = file.readlines()

        # Extract cluster data
        cluster_data_start = False
        cluster_info = []
        total_clusters = 0
        
        for line in content:
            if "CLUSTER DATA:" in line:
                cluster_data_start = True
                continue
            if cluster_data_start:
                if "Total Clusters" in line:
                    total_clusters = int(re.search(r'\d+', line).group())
                elif "Cluster" in line:
                    match = re.search(r'Cluster \d+ - Leader Cells: (\d+), Follower Cells: (\d+), Total Cells: (\d+)', line)
                    if match:
                        leader_cells = int(match.group(1))
                        follower_cells = int(match.group(2))
                        total_cells = int(match.group(3))
                        cluster_info.append([leader_cells, follower_cells, total_cells])
        
        # Append data to records
        for idx, cluster in enumerate(cluster_info, start=1):
            records.append([
                contact_energy, chemotaxis_lambda, proliferative_probability,
                total_clusters, idx, cluster[0], cluster[1], cluster[2]
            ])

# Create a DataFrame
columns = ["Contact Energy", "Chemotaxis Lambda", "Proliferative Probability",
           "Total Clusters", "Cluster ID", "Leader Cells", "Follower Cells", "Total Cells"]
df = pd.DataFrame(records, columns=columns)

# Sort DataFrame by parameter values in ascending order
df = df.sort_values(by=["Contact Energy", "Chemotaxis Lambda", "Proliferative Probability"])

# Save DataFrame to CSV
output_csv_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/Results/Data/cluster_data_output.csv"
df.to_csv(output_csv_path, index=False)

print(f"Cluster data saved to: {output_csv_path}")
```

```python
import glob
import os
import re
import pandas as pd

# Specify the path to the folder containing the data
base_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/TumorInvasionParameterScan"
file_pattern = "**/cclc_math_path/data_*.txt"  # Use glob pattern for all txt files in nested subfolders

# Gather all file paths matching the pattern
file_paths = glob.glob(os.path.join(base_path, file_pattern), recursive=True)

# Dictionary to store extracted values
records = []

# Loop through each file, load data, and collect the required values
for file_path in file_paths:
    # Extract parameters from the filename using regex
    match = re.search(r'data_(-?[\d.]+)_([\d.]+)_([\d.]+)', os.path.basename(file_path))
    iteration_match = re.search(r'scan_iteration_(\d+)', file_path)

    if match and iteration_match:
        # Extract parameter values
        contact_energy = float(match.group(1).rstrip('.'))
        chemotaxis_lambda = float(match.group(2).rstrip('.'))
        proliferative_probability = float(match.group(3).rstrip('.'))
        scan_iteration = int(iteration_match.group(1))  # Extract scan_iteration as integer

        # Initialize dictionary for each file
        data_entry = {
            "Scan Iteration": scan_iteration,
            "Contact Energy": contact_energy,
            "Chemotaxis Lambda": chemotaxis_lambda,
            "Proliferative Probability": proliferative_probability,
            "Invasive Area": None,
            "Infiltrative Area": None,
            "Single Defects": None,
            "Fingers": None,
            "Detached Cells": None,
            "Clusters": None
        }

        # Read the txt file to get all relevant metrics
        with open(file_path, 'r') as file:
            for line in file:
                if "Invasive Area" in line:
                    try:
                        data_entry["Invasive Area"] = float(line.split(":")[1].strip())
                    except ValueError:
                        print(f"Warning: Could not extract valid Invasive Area from {file_path}")

                elif "Infiltrative Area" in line:
                    try:
                        data_entry["Infiltrative Area"] = float(line.split(":")[1].strip())
                    except ValueError:
                        print(f"Warning: Could not extract valid Infiltrative Area AUC from {file_path}")

                elif "Single Defects" in line:
                    try:
                        data_entry["Single Defects"] = float(line.split(":")[1].strip())
                    except ValueError:
                        print(f"Warning: Could not extract valid Single Defected Cells from {file_path}")

                elif "Branches" in line:
                    try:
                        data_entry["Fingers"] = float(line.split(":")[1].strip())
                    except ValueError:
                        print(f"Warning: Could not extract valid Branches from {file_path}")

                elif "Detached Cells" in line:
                    try:
                        data_entry["Detached Cells"] = float(line.split(":")[1].strip())
                    except ValueError:
                        print(f"Warning: Could not extract valid Final Count of Detached Cells from {file_path}")
                        
                elif "Total Clusters" in line:
                    try:
                        data_entry["Clusters"] = float(line.split(":")[1].strip())
                    except ValueError:
                        print(f"Warning: Could not extract valid Total Clusters from {file_path}")

        # Append extracted values
        records.append(data_entry)

# Convert to a DataFrame
df = pd.DataFrame(records)

# Sort ONLY by Scan Iteration
df = df.sort_values(by=["Scan Iteration"], ascending=True)

# Save to CSV
output_csv_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/Results/Data/InvasionMetrics/Combined_Metrics.csv"
df.to_csv(output_csv_path, index=False)

print(f"Results saved successfully to {output_csv_path}")
```

### PHENOTYPE CLASSIFICATION

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/Results/Data/InvasionMetrics/Combined_Metrics.csv"
df = pd.read_csv(file_path)

# Define the phenotype classification function
def classify_phenotype(row):
    invasive_area = row["Invasive Area"]
    infiltrative_area = row["Infiltrative Area"]
    branches = row["Fingers"]
    single_defect = row["Single Defects"]
    clusters = row["Clusters"]
    
    if (invasive_area == infiltrative_area and (branches == 0 or branches == 1)) and single_defect == 0 and clusters == 0:
        return "No invasion"
    elif (invasive_area < infiltrative_area and branches == 0 and single_defect > 0 and clusters == 0):
        return "Single cell invasion"
    elif (invasive_area == infiltrative_area and branches > 1 and single_defect == 0 and clusters == 0):
        return "Bulk invasion"
    elif (invasive_area < infiltrative_area and branches > 1 and single_defect > 0 and clusters >= 0):
        return "Multimodal invasion"
    else:
        return "Unclassified"

# Apply the classification function
df["Phenotype"] = df.apply(classify_phenotype, axis=1)

# Remove rows with 'Unclassified' phenotype
df = df[df["Phenotype"] != "Unclassified"].reset_index(drop=True)

# Keep Scan Iteration and all relevant columns
columns_to_keep = [
    "Scan Iteration", "Contact Energy", "Chemotaxis Lambda", "Proliferative Probability",
    "Invasive Area", "Infiltrative Area", "Single Defects", "Fingers", "Detached Cells", "Clusters", "Phenotype"
]
df = df[columns_to_keep]

# Save the updated DataFrame to a new CSV
output_file_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/Results/Data/Phenotype/phenotype_classification.csv"
df.to_csv(output_file_path, index=False)

print(f"Phenotype classification saved to: {output_file_path}")

# Plot the distribution of phenotypes
phenotype_counts = df["Phenotype"].value_counts().sort_index()

plt.figure(figsize=(8, 5))
phenotype_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Distribution of Classified Invasion Phenotypes")
plt.xlabel("Phenotype")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the CSV file
file_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/Results/Data/InvasionMetrics/Combined_Metrics.csv"
df = pd.read_csv(file_path)

# Define the phenotype classification function
def classify_phenotype(row):
    branches = row["Fingers"]
    single_defect = row["Single Defects"]
    clusters = row["Clusters"]
    
    if (branches == 0 or branches == 1) and single_defect == 0 and clusters == 0:
        return "No invasion"
    elif (branches == 0 or branches == 1) and single_defect > 0 and clusters == 0:
        return "Single cell invasion"
    elif branches > 0 and single_defect == 0 and clusters == 0:
        return "Bulk invasion"
    elif branches > 0 and single_defect >= 0 and clusters >= 0:
        return "Multimodal invasion"
    else:
        return "Unclassified"

# Apply phenotype classification
df["Phenotype"] = df.apply(classify_phenotype, axis=1)

# Keep only parameter + phenotype
param_cols = ["Contact Energy", "Chemotaxis Lambda", "Proliferative Probability", "Phenotype"]
df_filtered = df[param_cols]

# Group by parameter combination and phenotype count
phenotype_counts = df_filtered.groupby(["Contact Energy", "Chemotaxis Lambda", "Proliferative Probability", "Phenotype"]).size().reset_index(name="Count")

# Normalize to get probability
phenotype_probs = phenotype_counts.copy()
phenotype_probs["Total"] = phenotype_probs.groupby(["Contact Energy", "Chemotaxis Lambda", "Proliferative Probability"])["Count"].transform("sum")
phenotype_probs["Probability"] = phenotype_probs["Count"] / phenotype_probs["Total"]

# Keep only necessary columns
phenotype_probs = phenotype_probs[["Contact Energy", "Chemotaxis Lambda", "Proliferative Probability", "Phenotype", "Probability"]]

# Save to CSV
output_file_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/Results/Data/Phenotype/phenotype_probability_map.csv"
phenotype_probs.to_csv(output_file_path, index=False)

print(f"Phenotype probability map saved to: {output_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "/Users/sheriffakeeb/Dropbox (Personal)/CollectiveInvasion/Results/Data/Phenotype/phenotype_classification.csv"
df = pd.read_csv(file_path)

# --- 1. Plot the overall phenotype distribution ---
phenotype_counts = df['Phenotype'].value_counts()

plt.figure(figsize=(8, 6))
phenotype_counts.plot(kind='bar')
plt.xlabel('Phenotype')
plt.ylabel('Number of Simulations')
plt.title('Distribution of Phenotypes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- 2. Analyze Bulk invasion (Fingers) ---
bulk_df = df[df['Phenotype'] == 'Bulk invasion']

bulk_1_finger = (bulk_df['Fingers'] == 1).sum()
bulk_2_fingers = (bulk_df['Fingers'] == 2).sum()
bulk_3plus_fingers = (bulk_df['Fingers'] >= 3).sum()

bulk_summary = {
    '1 finger': bulk_1_finger,
    '2 fingers': bulk_2_fingers,
    '≥3 fingers': bulk_3plus_fingers
}

# --- 3. Analyze Single Cell invasion (Defectors) ---
single_cell_df = df[df['Phenotype'] == 'Single cell invasion']

single_1_defector = (single_cell_df['Single Defects'] == 1).sum()
single_2_defectors = (single_cell_df['Single Defects'] == 2).sum()
single_3plus_defectors = (single_cell_df['Single Defects'] >= 3).sum()

single_cell_summary = {
    '1 defector': single_1_defector,
    '2 defectors': single_2_defectors,
    '≥3 defectors': single_3plus_defectors
}

# --- 4. Plot pie chart for Bulk invasion (Fingers) ---
plt.figure(figsize=(7, 6))
plt.pie(
    bulk_summary.values(),
    labels=bulk_summary.keys(),
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Bulk Invasion: Finger Distribution')
plt.tight_layout()
plt.show()

# --- 5. Plot pie chart for Single Cell invasion (Defectors) ---
plt.figure(figsize=(7, 6))
plt.pie(
    single_cell_summary.values(),
    labels=single_cell_summary.keys(),
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Single Cell Invasion: Defector Distribution')
plt.tight_layout()
plt.show()

# --- 6. Print out the exact counts if you want ---
print("Bulk Invasion Summary:", bulk_summary)
print("Single Cell Invasion Summary:", single_cell_summary)
```
